```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load data
dataset = pd.read_csv('Position_Salaries.csv')
print(dataset)

x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
#training the linear regression model on the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
#training the polynomial linear regression model on the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
#visualizing the linear regression results
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color = 'blue')
plt.title("Truth of bluff (Linear Regression)")
plt.xlabel("Position Label")
plt.ylabel("Salary")
```

```python
#visualizing the polynomial linear regression results
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)),color = 'blue')
plt.title("Truth of bluff (Linear Regression)")
plt.xlabel("Position Label")
plt.ylabel("Salary")
```

```python
#visualizing the polynomial linear regression results with smoother curve
x_grid  = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)),color = 'blue')
plt.title("Truth of bluff (Linear Regression)")
plt.xlabel("Position Label")
plt.ylabel("Salary")
```

```python
#predicting new result with linear regression
lin_reg.predict([[6.5]])
```

```python
#predicting new result with polynomial linear regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
